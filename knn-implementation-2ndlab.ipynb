```python
from math import sqrt
```

```python
def euclidean_distance(x, y):
    distance = 0
    length = len(y)
    for i in range(length):
        distance += (x[i] - y[i]) ** 2
    return sqrt(distance)
    
""" 
the double-star operator (**) is used for exponentiation,
where one number is raised to the power of another.  
2 ** 3, which would return 8.
"""
    
        
```

```python
# Locate the most similar neighbors
def get_neighbors(train, test_row, num_neighbors):
    distances = list()
    #print(train)
    #train_row = [7, 7, 0] # 0 indicates labeling(good or bad)
    #test_row = [3, 7]
    for train_row in train:
        dist = euclidean_distance(train_row, test_row) # ([7,7], [3,7])
        distances.append((train_row, dist))
        #[[3,7], 4]
    #print(distances)
    distances.sort(key= lambda tup: tup[1])
    #print(distances)
    neighbors = list()
    for i in range(num_neighbors):
        neighbors.append(distances[i][0])
    #print(neighbors)
    return neighbors
```

```python
# Make a classificatin prediction with neighbors
def predict_classification(train, test_row, num_neighbors):
    neighbors = get_neighbors(train, test_row, num_neighbors)
    output_values = [row[-1] for row in neighbors]
    print(output_values)
    #return value which has majority occurance
    print(set(output_values)) # set: re-sorting if the previous sort is incomplete
    prediction = max(set(output_values), key=output_values.count)
    return prediction
```

```python
labels = ["Bad", "Good"] # 0 bad, 1 good
dataset = [
    [7, 7, 0],
    [7, 4, 0],
    [3, 4, 1],
    [1, 4, 1]
]
```

```python
test_data = [3, 7]

k = int(input("Enter k vale = "))
"""
dataset: contains samples data
test_data: to test KNN learning model against given data set
k: number of neighbours
"""
#to predict class for test data
prediction = predict_classification(dataset, test_data, k)
#('Expected %d, Got %d.' %(dataset[0][-1], prediction))
print('Got %s.' % (labels[prediction]))
```
